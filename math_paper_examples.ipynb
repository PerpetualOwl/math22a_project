```python
from matplotlib import image
from matplotlib import pyplot as plt
import numpy as np
image = image.imread('/Users/howardhuang/Downloads/flowers.jpg')
print("Original")
plt.imshow(image)
plt.show()
```

```python
blurred_rgb = np.copy(image)

for k in range(3):
  for i in range(2, image.shape[0] - 2):
    for j in range(2, image.shape[1] - 2):
      val = 0
      div = 0
      for x in range(-2, 3):
        for y in range(-2, 3):
          val += image[i + x][j + y][k] * (5 - abs(x) - abs(y))
          div += (5 - abs(x) - abs(y))
      val = val // div
      blurred_rgb[i][j][k] = int(val)

print("Blurred")
plt.imshow(blurred_rgb)
plt.show()
```

```python
grayscale = np.zeros((image.shape[0], image.shape[1]))

for k in range(3):
  for i in range(2, image.shape[0] - 2):
    for j in range(2, image.shape[1] - 2):
      grayscale[i][j] += image[i][j][k] // 3

print("Grayscale")
plt.imshow(grayscale, cmap=plt.get_cmap('gray'), vmin=0, vmax=256)
plt.show()
```

```python
edge_gray1 = np.copy(grayscale)

for i in range(1, grayscale.shape[0] - 1):
  for j in range(1, grayscale.shape[1] - 1):
    val = 0
    for x in range(-1, 2):
      for y in range(-1, 2):
        val += grayscale[i + x][j + y] * x
    edge_gray1[i][j] = int(val)

print("Edge Detection (horizontal)")
plt.imshow(edge_gray1, cmap=plt.get_cmap('gray'), vmin=0, vmax=256)
plt.show()
```

```python
edge_gray2 = np.copy(grayscale)

for i in range(1, grayscale.shape[0] - 1):
  for j in range(1, grayscale.shape[1] - 1):
    val = 0
    for x in range(-1, 2):
      for y in range(-1, 2):
        val += grayscale[i + x][j + y] * y
    edge_gray2[i][j] = int(val)

print("Edge Detection (vertical)")
plt.imshow(edge_gray2, cmap=plt.get_cmap('gray'), vmin=0, vmax=256)
plt.show()
```

```python
gray_sharpen = np.copy(grayscale)

for i in range(1, grayscale.shape[0] - 1):
  for j in range(1, grayscale.shape[1] - 1):
    val = 0
    for x in range(-1, 2):
      for y in range(-1, 2):
        val += grayscale[i + x][j + y] * (-1)
    val += grayscale[i][j] * 10
    gray_sharpen[i][j] = int(val)

print("Sharpen")
plt.imshow(gray_sharpen, cmap=plt.get_cmap('gray'), vmin=0, vmax=256)
plt.show()
```

```python
color_shift = np.copy(image)

for k in range(3):
  for i in range(image.shape[0]):
    for j in range(image.shape[1]):
      color_shift[i][j][k] = image[i][j][(k + 1) % 3]

print("Color Shift")
plt.imshow(color_shift)
plt.show()
```
